```r
library(MASS)
# 사용할 데이터에 대한 설명
?Boston
```

```r
# 데이터 확인
Boston
```

```r
# 데이터를 훈련 세트와 테스트 세트로 분할하기 위한 인덱스를 생성
idx <- sample(1:nrow(Boston), size=nrow(Boston) * 0.7, replace=F)
Boston_train <- Boston[idx, ]   # 훈련용 데이터 세트 70% 354개 행
Boston_test <- Boston[-idx, ]   # 평가용 데이터 세트 30% 152개 행
dim(Boston_train); dim(Boston_test)
```

### 다중회귀분석

```r
# medv를 목표변수 나머지 변수를 독립변수로 다중회귀분석 진행
lm.fit <- lm(medv~., data=Boston_train)
# 모델에 대한 요약 출력
summary(lm.fit)
```

```r
# 단계적 선택변수 선택 (양방향)
lm.fit2 <- step(lm.fit, method="both")
# 모델에 대한 요약 출력
summary(lm.fit2)
```

```r
# 평가데이터로 예측 진행
lm.yhat2 <- predict(lm.fit2, newdata=Boston_test)
mean((lm.yhat2-Boston_test$medv)^2) # MSE 계산
```

### 의사결정트리

```r
# tree 라이브러리 사용
library(tree)
# medv를 목표변수 나머지 변수를 독립변수로 의사결정트리 진행
tree.fit <- tree(medv~., data=Boston_train)
# 모델에 대한 요약 출력
summary(tree.fit)
```

```r
# 의사결정트리 시각화
plot(tree.fit)
# 시각화된 차트의 노드에 레이블 추가
text(tree.fit, pretty=0)
```

```r
# 평가 데이터로 예측 결과 생성
tree.yhat <- predict(tree.fit, newdata=Boston_test)
mean((tree.yhat - Boston_test$medv)^2) # MSE 계산
```

```r
# rpart 라이브러리 사용
library(rpart)
# medv를 목표변수 나머지 변수를 독립변수로 의사결정트리 진행
rpart.fit <- rpart(medv~., data=Boston_train)
# 모델에 대한 요약 출력
summary(rpart.fit)
```

```r
# rpart 의사결정트리 시각화 라이브러리 가져오기
library(rpart.plot)
# 의사결정트리 시각화
rpart.plot(rpart.fit, digits=3, type=0, extra=1, fallen.leaves=F, cex=1)
```

```r
# 평가 데이터로 예측 결과 생성
rpart.yhat <- predict(rpart.fit, newdata=Boston_test)
mean((rpart.yhat - Boston_test$medv)^2) # MSE 계산
```

```r
# 정규화 함수 작성 (MinMaxScaler)
normalize <- function(x){ return((x-min(x))/(max(x)-min(x))) }
# 학습데이터에 정규화 함수 적용
Boston_train_norm <- as.data.frame(sapply(Boston_train, normalize))
# 평가데이터에 정규화 함수 적용
Boston_test_norm <- as.data.frame(sapply(Boston_test, normalize))
```

```r
library(nnet)
# medv를 목표변수 나머지 변수를 독립변수로 인공신경망 진행 (은닉층 5개)
nnet.fit <- nnet(medv~., data=Boston_train_norm, size=5)
# 평가 데이터로 예측 결과 생성
nnet.yhat <- predict(nnet.fit, newdata=Boston_test_norm, type="raw")
mean((nnet.yhat - Boston_test_norm$medv)^2) # MSE 계산
```

```r
library(neuralnet)
# 인공신경망 적합 (은닉층 5개)
neural.fit <- neuralnet(
    medv~crim+zn+indus+chas+nox
    +rm+age+dis+rad+tax+ptratio
    +black+lstat,
    data=Boston_train_norm,
    hidden=5
)
# 평가데이터로 예측 결과 생성
neural.results <- compute(neural.fit, Boston_test_norm[1:13])
neural.yhat <- neural.results$net.result
mean((neural.yhat - Boston_test_norm$medv)^2) # MSE 계산
```
