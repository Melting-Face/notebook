### 패키지 가져오기 및 데이터확인

```r
library(arules)
library(arulesViz)
data(Groceries)
```

```r
data(package = "arules")
```

```r
Groceries
inspect(Groceries)
```

```r
summary(Groceries)
```

```r
sort(itemFrequency(Groceries, type="absolute"), decreasing=T)
# 상대빈도
round(sort(itemFrequency(Groceries, type="relative"), decreasing=T), 3)
# 절대빈도 시각화
itemFrequencyPlot(Groceries, topN=10, type="absolute")
# 상대빈도 시각화
itemFrequencyPlot(Groceries, topN=10, type="relative")
```

```r
# apriori 연관성분석
apriori(Groceries)
# 신뢰도 0.5, 지지도 0.005로 변경 후
result_rules <- apriori(
    Groceries,
    parameter=list(support=0.005, confidence=0.5, minlen=2)
)
```

```r
summary(result_rules)
```

```r
inspect(result_rules[1:5])
```

```r
rules_lift<-sort(result_rules, by="lift", decreasing=T)
inspect(rules_lift[1:5])
```

```r
rules_conf<-sort(result_rules, by="confidence", decreasing=T)
inspect(rules_conf[1:5])
```

```r
milk_rule <- subset(rules_lift, items %in% "whole milk")
milk_rule
inspect(milk_rule)
```

```r
rhs.milk_rule <- subset(rules_lift, rhs %in% "whole milk")
rhs.milk_rule
inspect(rhs.milk_rule[1:5])
```

```r
#  whole milk를 결과(RHS)로 하는 연관 규칙 생성
wholemilk_rule <- apriori(
    Groceries,
    parameter=list(
        support=0.005,
        confidence=0.5,
        minlen=2
    ),
    appearance=list(
        default="lhs",
        rhs="whole milk"
    )
)
# 생성된 규칙을 향상도 기준으로 정렬
wholemilk_rule <- sort(
    wholemilk_rule,
    by="lift",
    decreasing=T
)
# 상위 5개 규칙 출력
inspect(wholemilk_rule[1:5])
```

```r
library(arulesViz)
plot(wholemilk_rule[1:10], method="graph", measure="lift", shading="confidence")
```
